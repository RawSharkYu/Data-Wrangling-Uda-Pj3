```python
import pandas as pd
```

### First analyze data to get insights.

```python
df = pd.read_csv("twitter_archive_master.csv", encoding='utf-8')
```

```python
df
```

```python
df.info()
```

1.Count the source of tweets.

```python
df["source"].value_counts()
```

Mostly the source is the "Twitter for iPhone"

2.The correaltionship between favorite_count and retweet_count

```python
df[["favorite_count","retweet_count"]].corr()
```

The favorite_count and retweet_count are highly correlated.

3.The correaltionship between favorite_count and rating_numerator

```python
df[["favorite_count","rating_numerator"]].corr()
```

To my surprise, there is no obvious correlationship between favorite_count and rating_numerator. That means, the dogs get high scores by "WeRateDogs" do not usually get followers' favorites.

### Then, visulize data to get more information.
